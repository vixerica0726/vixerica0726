```python
# Importação dos dados
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 1. Análise do faturamento

```python
import matplotlib.pyplot as plt

# Calculando o faturamento total de cada loja
faturamento_loja1 = loja['Preço'].sum() + loja['Frete'].sum()  # Soma o preço e o frete de todos os produtos da Loja 1
faturamento_loja2 = loja2['Preço'].sum() + loja2['Frete'].sum()  # Soma o preço e o frete de todos os produtos da Loja 2
faturamento_loja3 = loja3['Preço'].sum() + loja3['Frete'].sum()  # Soma o preço e o frete de todos os produtos da Loja 3
faturamento_loja4 = loja4['Preço'].sum() + loja4['Frete'].sum()  # Soma o preço e o frete de todos os produtos da Loja 4

# Criando gráfico de barras para visualizar o faturamento
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']  # Definindo o nome das lojas
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]  # Lista com o faturamento das lojas

plt.bar(lojas, faturamento, color='skyblue')  # Criando o gráfico de barras
plt.xlabel('Lojas')  # Rótulo para o eixo x
plt.ylabel('Faturamento (R$)')  # Rótulo para o eixo y
plt.title('Faturamento Total por Loja')  # Título do gráfico
plt.show()  # Exibe o gráfico
```

# 2. Vendas por Categoria

```python
# Juntando os dados das quatro lojas em um único DataFrame
# Isso facilita a análise geral dos dados
todas_lojas = pd.concat([loja, loja2, loja3, loja4])

# Contando quantos produtos foram vendidos por categoria
# Ou seja, quantas vezes cada categoria aparece na coluna "Categoria do Produto"
vendas_categoria = todas_lojas['Categoria do Produto'].value_counts()

# Exibindo o resultado
print("Quantidade de vendas por categoria de produto:\n")
print(vendas_categoria)

# Criando gráfico de barras para visualizar melhor
vendas_categoria.plot(kind='bar', color='skyblue')
plt.title('Vendas por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
# Média de avaliação das lojas
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Gráfico de média de avaliação
avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

plt.bar(lojas, avaliacoes, color='lightgreen')
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.title('Média de Avaliação das Lojas')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos


```python
# Produto mais vendido
produto_mais_vendido_loja1 = loja['Produto'].mode()[0]
produto_mais_vendido_loja2 = loja2['Produto'].mode()[0]
produto_mais_vendido_loja3 = loja3['Produto'].mode()[0]
produto_mais_vendido_loja4 = loja4['Produto'].mode()[0]

# Gráfico de produtos mais vendidos
produtos = [produto_mais_vendido_loja1, produto_mais_vendido_loja2, produto_mais_vendido_loja3, produto_mais_vendido_loja4]

plt.bar(lojas, produtos, color='lightblue')
plt.xlabel('Lojas')
plt.ylabel('Produto Mais Vendido')
plt.title('Produto Mais Vendido por Loja')
plt.show()
```

# 5. Frete Médio por Loja


```python
# Frete médio das lojas
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Gráfico de frete médio
fretes = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

plt.bar(lojas, fretes, color='lightyellow')
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.title('Frete Médio por Loja')
plt.show()
```
